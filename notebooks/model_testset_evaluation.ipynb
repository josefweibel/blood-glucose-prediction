```python
import os
if os.getcwd().endswith('notebooks'):
  os.chdir('..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.training import investigate_predictions, calculate_metrics, load_config, DATA_TIME_INTERVAL
```

```python
model_name = '053-pred_horizon_60' # change
config = load_config(model_name)
```

```python
model_results = []
for i in range(config['repetitions']):
  X, y_preds, y_trues = investigate_predictions(model_name, iteration=i, split='test')
  for j in [3, 6, 9, 12]:
    model_results.append({'horizon': f'{j * 5} min', **calculate_metrics(y_preds[:, j], y_trues[:, j])})
  model_results.append({'horizon': '5-60 min (5 min steps)', **calculate_metrics(y_preds, y_trues)})
model_results = pd.DataFrame(model_results)
model_results
```

```python
changed_model_results = model_results.copy()
changed_model_results['mape'] = changed_model_results['mape'] * 100
means = changed_model_results.groupby('horizon').mean().round(2)
stds = changed_model_results.groupby('horizon').std().round(2)

table = means.astype(str).add(' Â± ').add(stds.astype(str))
table.columns = table.columns.str.upper()
table = table.rename(columns={'GMSE': 'gMSE', 'GRMSE': 'gRMSE'})
table = table[['RMSE', 'MAPE', 'gRMSE']]
table.index.name = None
table
```
